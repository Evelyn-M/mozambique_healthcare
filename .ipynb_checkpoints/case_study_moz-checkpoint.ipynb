```python
gdf_health = HNDFileQuery.retrieve_cis('healthcare')
```

```python
gdf_hospital = gdf_health[gdf_health.amenity=='hospital']
```

```python
# plot results
ax = gdf_hospital.set_crs(epsg=4326).to_crs(epsg=3857).plot(figsize=(15, 15), alpha=1, markersize=5, color='blue', 
                    edgecolor='blue', label='roads HND')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles, loc='upper left')
ax.set_title('Hospitals Mozambique', fontsize=25)
ctx.add_basemap(ax)
plt.show()
```

```python
import climada.util.plot as u_plot
import cartopy.crs as ccrs
from cartopy.io import shapereader
import geopandas
shp_file = shapereader.natural_earth(resolution='10m', category='cultural',
                                     name='admin_0_countries')
df = geopandas.read_file(shp_file)
df = df[df['ADMIN']=='Mozambique']

extent = (28, 60, -35.20000076293945, -3.700000047683716)
proj = ccrs.PlateCarree(central_longitude=0.5 * (60 + 26))
_, axis, _ = u_plot.make_map(proj=proj)
axis.set_extent(extent)
u_plot.add_shapes(axis)


for geometry in df.geometry:
    axis.add_geometries([geometry], crs=ccrs.PlateCarree(), facecolor='gray',
                        edgecolor='dimgray')
    
gdf_hospital = gdf_hospital

for geometry in gdf_hospital.geometry:
    axis.add_geometries([geometry], crs=ccrs.PlateCarree(), facecolor='red',
                        edgecolor='dimgray')

tracks.subset({'sid':'2019063S18038'}).plot(axis=axis)
```
